```python
import time
import datetime
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService

data_de_hoje = datetime.date.today()

precos_carrefour = []
```

```python
def check_all_stores(j, j_max, tabela_precos):
    carrefour = "https://mercado.carrefour.com.br/s?q=picanha&sort=score_desc&page=0"
    lojas = []
    loja_atual = 0
    while j <= j_max: 
        driver = webdriver.Chrome(service=ChromeService(
        ChromeDriverManager().install()))
        driver.get(carrefour)

        time.sleep(2)

        driver.find_element(By.CSS_SELECTOR, "div[aria-label='Retire na Loja']").click()

        time.sleep(2)

        select = Select(driver.find_element(By.CSS_SELECTOR, "#selectCity"))

        if select.options[j].is_enabled():
            select.select_by_index(j)
        else:
            j += 1
            driver.quit()
            continue
        
        lojas = driver.find_element(By.CLASS_NAME, "max-h-52.overflow-y-scroll.snap-y.pb-4").find_elements(By.TAG_NAME, "article")

        botao = lojas[loja_atual]
        loja_atual += 1
        time.sleep(2)
        try:
            loja, _, cidade, _ = botao.text.split("\n")
            botao.click()
        except:
            botao = lojas[loja_atual]
            loja, _, cidade, _ = botao.text.split("\n")
            botao.click()

        time.sleep(5)
        for _ in range(10):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        divs_carrefour = driver.find_elements(By.CSS_SELECTOR, "section[data-product-card-content='true']")

        for div in divs_carrefour:
            dtxt = div.text
            if len(dtxt) > 0:
                tabela_precos.append(dtxt + " " + loja + "" + cidade)

        if len(lojas) == (loja_atual):
            loja_atual = 0
            j += 1

    driver.quit()
```

```python
carrefour = "https://mercado.carrefour.com.br/s?q=picanha&sort=score_desc&page=0"
driver = webdriver.Chrome(service=ChromeService(
        ChromeDriverManager().install()))
driver.get(carrefour)
time.sleep(2)
driver.find_element(By.CSS_SELECTOR, "div[aria-label='Retire na Loja']").click()
time.sleep(2)
select = Select(driver.find_element(By.CSS_SELECTOR, "#selectCity"))
cidades = [opt.text for opt in select.options]
driver.quit()
print(f"Ao todo são {len(cidades) - 2} cidades disponíveis")
```

```python
# ajustando a tabela de preços
df = {
    "produto":[],
    "preco":[],
    "desconto":[],
    "loja":[],
    "cidade/UF":[],
    'estado':[],
    "data": concat_bacalhau["data"].tolist()
}

for i in concat_bacalhau[0].tolist():
    info = i.split("\n")

    if len(info) == 2:


        produto = info[0]

        preco_loja, cidade_estado = info[1].split(", ")

        cidade, estado = cidade_estado.split(" - ")

        preco_loja = preco_loja.lstrip("R$ ")
        space = preco_loja.find(" ")
        preco = preco_loja[:space]
        loja = preco_loja[space+1:]

        desconto = 0

    elif len(info) == 3:


        produto = info[0]
        preco = info[1].lstrip("R$ ")

        loja, cidade_estado = info[2].split(", ")

        cidade, estado = cidade_estado.split(" - ")

        _, loja = loja.split("/kg ")
        desconto = 0

    elif len(info) == 4:
        desconto_produto = info[0].split("%")

        if len(desconto_produto) == 2:
            desconto = int(desconto_produto[0])
            produto = info[1]
        else:
            desconto = int(info[2][1:].rstrip("%"))
            produto = desconto_produto[0]

        preco_loja, cidade_estado = info[3].split(", ")

        cidade, estado = cidade_estado.split(" - ")

        preco_loja = preco_loja.lstrip("R$ ")
        space = preco_loja.find(" ")
        preco = preco_loja[:space]
        loja = preco_loja[space+1:]

    elif len(info) == 5:

        desconto = int(info[0].split("%")[0])
        produto = info[1]
        preco = info[3].lstrip("R$ ")

        loja, cidade_estado = info[4].split(", ")

        cidade, estado = cidade_estado.split(" - ")

        _, loja = loja.split("/kg ")

    elif len(info) == 6:
        desconto = int(info[4][1:].rstrip("%"))
        produto = info[1]

        preco_loja, cidade_estado = info[5].split(", ")

        cidade, estado = cidade_estado.split(" - ")

        preco_loja = preco_loja.lstrip("R$ ")
        space = preco_loja.find(" ")
        preco = preco_loja[:space]
        loja = preco_loja[space+1:]


    preco = preco.replace(",", ".")
    preco = float(preco)
    df["produto"].append(produto)
    df["preco"].append(preco)
    df["desconto"].append(desconto)
    df["loja"].append(loja)
    df["cidade/UF"].append(cidade+'/'+estado)
    df["estado"].append(estado)


df_precos = pd.DataFrame().from_dict(df)
df_precos = df_precos.drop_duplicates()
df_precos.head(5)
```

```python
precos_carrefour_prod = precos_carrefour[0]
for i in range(len(precos_carrefour_prod)):
    precos_carrefour_prod[i] = precos_carrefour_prod[i].split("\n")
#explodir dados
precos_carrefour_prod = pd.DataFrame(precos_carrefour_prod.tolist(), index=precos_carrefour_prod.index)
#renomear colunas
precos_carrefour_prod = precos_carrefour_prod.rename(columns={0: "Produto", 1: "Preco",2:"Desconto", 3:"Preco c/Desc", 4: "Loja"})

#alterar a posiçao dos dados das lojas que estavam no desconto
precos_carrefour_prod['Loja'].fillna(value=precos_carrefour_prod['Desconto'], inplace=True)
# eliminar os dados da loja da coluna desconto
from cmath import nan
precos_carrefour_prod.Desconto[precos_carrefour_prod.Desconto == precos_carrefour_prod['Loja']] = nan

precos_carrefour_prod['Loja2'] = precos_carrefour_prod['Loja'].loc[precos_carrefour_prod['Loja'].str[:4] == 'ADIC']
precos_carrefour_prod['Loja2'].fillna(value=precos_carrefour_prod['Preco c/Desc'].loc[precos_carrefour_prod['Preco c/Desc'].str[:4] == 'ADIC'],inplace=True)

precos_carrefour_prod['Preco c/Desc2'] = precos_carrefour_prod['Loja'].loc[precos_carrefour_prod['Loja'].str[:2] == 'R$']
precos_carrefour_prod['Preco c/Desc2'].fillna(value=precos_carrefour_prod['Preco c/Desc'].loc[precos_carrefour_prod['Preco c/Desc'].str[:2] == 'R$'],inplace=True)

precos_carrefour_prod = precos_carrefour_prod[['Produto', 'Preco', 'Desconto', 'Preco c/Desc2', 'Loja2']]
precos_carrefour_prod = precos_carrefour_prod.rename(columns={'Preco c/Desc2':"Preco c/Desc", 'Loja2': "Loja"})

#Ajuste no Preço
precos_carrefour_prod['Preco'] = precos_carrefour_prod['Preco'].str.replace(r'\D+','', regex=True).astype(float)
precos_carrefour_prod['Preco'] = precos_carrefour_prod['Preco']/100

#Ajuste na Loja - Separar Loja de Cidade
precos_carrefour_prod['Loja'] = precos_carrefour_prod['Loja'].str.replace('ADICIONAR ','', regex=True)
precos_carrefour_prod[['Loja', 'Cidade - UF']] = precos_carrefour_prod['Loja'].str.split(', ', expand=True)
```
